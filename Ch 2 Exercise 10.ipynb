```python
#item d
#scatter plot of crim vs each column. 
crim_column = Boston['crim']

# Specify the dimensions of the figure
fig, axes = plt.subplots(nrows=3, ncols=5, figsize=(12, 6))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Create individual scatter plots for 'mpg' versus other columns
for i, column in enumerate(Boston.columns[1:]):
    axes[i].scatter(Boston[column], crim_column)
    axes[i].set_title(f'Scatter plot of crim vs {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('crim')

# Adjust layout for better visibility
fig.tight_layout()

# Show the plot
plt.show()
```
